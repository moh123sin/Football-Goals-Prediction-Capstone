```python
# 01_download_epl_data.ipynb
# Step 1: imports and folders

import os
import pandas as pd
import requests

# assuming this notebook is inside football_capstone/notebooks
base_dir = os.path.dirname(os.getcwd())
data_dir = os.path.join(base_dir, "data")
os.makedirs(data_dir, exist_ok=True)

print("Base directory:", base_dir)
print("Data directory:", data_dir)
```

```python
# Step 2: define EPL CSV URLs for two seasons

season_urls = {
    "2223": "https://www.football-data.co.uk/mmz4281/2223/E0.csv",  # 2022-23
    "2324": "https://www.football-data.co.uk/mmz4281/2324/E0.csv",  # 2023-24
}

season_urls
```

```python
# Step 3: helper to download one season file

def download_season_csv(season_code, url, save_dir):
    print(f"Downloading season {season_code} from {url}")
    response = requests.get(url)
    response.raise_for_status()

    file_path = os.path.join(save_dir, f"raw_epl_{season_code}.csv")
    with open(file_path, "wb") as f:
        f.write(response.content)

    print("Saved:", file_path)
    return file_path
```

```python
# Step 4: download all seasons into data/ folder

season_files = []

for season_code, url in season_urls.items():
    path = download_season_csv(season_code, url, data_dir)
    season_files.append(path)

season_files
```

```python
# Step 5: load both CSVs and inspect shapes

df_2223 = pd.read_csv(season_files[0])
df_2324 = pd.read_csv(season_files[1])

print("2223 shape:", df_2223.shape)
print("2324 shape:", df_2324.shape)

df_2223.head()
```

```python
# Step 6: combine the two seasons into one DataFrame

df_all = pd.concat([df_2223, df_2324], ignore_index=True)
print("Combined shape:", df_all.shape)

combined_path = os.path.join(data_dir, "raw_epl_matches_2seasons.csv")
df_all.to_csv(combined_path, index=False)

print("Combined raw file saved at:", combined_path)
```
